```python
from langchain.embeddings import OpenAIEmbeddings
import pandas as pd
df = pd.read_csv('myData.csv')
print(df)
```

```python
embeddings = OpenAIEmbeddings()
```

**Convert Word To Embeddings**

```python
df['embedding'] = df['Words'].apply(lambda x: embeddings.embed_query(x))
df.to_csv('word_embeddings.csv')
```

**Load Embeddings**

```python
new_df = pd.read_csv('word_embeddings.csv')
print(new_df)
```

**Let's Get Embeddings For Our Text**

```python
our_text = 'mammal'
```

```python
text_embedding = embeddings.embed_query(our_text)
```

```python
print(f'Our embedding is {text_embedding}')
```

```python
from openai.embeddings_utils import cosine_similarity
df['similarity_score'] = df['embedding'].apply(lambda x: cosine_similarity(x, text_embedding))
df
```
